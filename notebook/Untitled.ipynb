```python
data = pd.read_csv("../data/raw/travel_dataset.csv")
data.head()
```

#### Standardizing column names

```python
cols = []
for col in data.columns:
    col = col.lower().replace(' ', '_')
    cols.append(col)
data.columns = cols
```

```python
data.head()
```

```python
data.to_csv("../data/clean/travel_dataset.csv", index=False)
```
